## Parametros e Argumentos

**Parametro** é a variavel definida na função, já o **argumento** é o valor enviado para a mesma

```python
def somar(a,b,c): # a,b e c são parametros
  print(a + b + c)

somar(13,23,12) # 13,23,12 são os argumentos
```

## Argumentos Nomeados e Não Nomeados

Quando chamamos uma função, passamos informações chamadas de **argumentos**

```python
# os argumentos não nomeados tem os seus valores passados pela ordem
somar(1,2,3)
```

Nesse caso, os argumentos foram passados em ordem e assim a soma foi feita

```python
# já os argumentos nomeados, informam explicitamente qual parametro está recebendo o seu valor
somar(a=1, b=2, c=3)
```

```python
# é possivel usar esse método para concatenar textos e misturar com argumentos não nomeados
def msg_somar(e,f):
  print(e+f)

msg_somar(e='olá ', f='mundo')
msg_somar(1, f=1) # Argumentos posicionais sempre vêm antes dos nomeados.
```

## Escopos

Escopo seria aonde uma variável pode ser acessada dentro de uma função, existem dois tipos de escopos principais, os **globais** e os **locais**

```python
# exemplo de um escopo global, um escopo local pode ser usado em qualquer lugar
x = 1
# o escopo local é criado dentro de uma função
def mostrar_x():
  x = 1
```

em uma função, pode-se usar a palavra "global",
que substitui o uso de uma variável local dentro de uma função.

```python
x = 1

def scopo():
  global x
  x = 10

scopo()
print(x) # o resultado é 10, pois foi mostrado o valor de x dentro da função escopo
```

## Empacotamento
### ARGS
a função args é uma **tupla** que permite receber uma quantidade variável de argumentos

```python
def soma(*args):
    total = 0

    for numero in args:
        total += numero

    return total

#passando diversos argumentos dentro de print
print(soma(10,10))
print(soma(10,10,10))
print(soma(10,10,10,10))
```

##Dicionários

#List, Dict e Set Comprehension
é uma forma de criar listas a partir de iteraveis, é uma forma de deixar o codigo mais legivel, economizando processos

```python
# sem list comprehension
lista = []
for numero in range(10):
  lista.append(numero)
print(lista)
```

```python
#com list comprehension
numero = 1
lista = [
    numero for numero in range(10)
]
print(lista)
```

```python
#sem dict comprehension
produtos = {
    "nome": "lapis",
    "preco": 1.50
}

for chave, valor in produtos.items():
  print(chave, valor)
```

```python
#com dict comprehension
produtos_origem = {
    "nome": "lapis",
    "preco": 1.50
}
produtos_novo = {chave: valor for chave, valor in produtos_origem.items()}
print(produtos_novo)
```

```python
#sem set comprehension
print(set(range(10)))
```

```python
#com set comprehension
s1 = {i for i in range(10)}
print(s1)
```

##Iteradores
um itarável, é um objeto que retorna __iter__, já um iterador implementa a função __next__, que retorna o proximo valor da sequencia e lança StopIteration quando esgotado.

```python
# exemplo de um iteravel

class	Contador:
  def	__init__(self,	limite):
    self.atual	=	0
    self.limite	=	limite

  def	__iter__(self):
    return	self

  def	__next__(self):
    if	self.atual	>=	self.limite:
      raise	StopIteration
    self.atual	+=	1
    return	self.atual

for	numero	in	Contador(5):
    print(numero)		#	1	2	3	4	5
```

##Geradores
Geradores são uma forma simples de criar iteradores usando a palavra chave yield. Ao inves de retornar todos os valores de uma vez, o gerador produz valores sob demanda, economizando memoria, essencial para trabalhar com grandes volumes de dados ou sequencias infinitas.

```python
# exemplo de um gerador
def	contador(limite):
  n	=	1
  while	n	<=	limite:
    yield	n
    n	+=	1

for	valor	in	contador(5):
    print(valor)
#	Expressão	geradora	(semelhante	a	list	comprehension,	porém	"preguiçosa")
quadrados_gen	=	(x**2	for	x	in	range(1000000))		#	não	ocupa	memória	imediatamente
```

```python
#	Gerador	infinito	de	números	de	Fibonacci
def	fibonacci():
  a,	b	=	0,	1
  while	True:
    yield	a
    a,	b	=	b,	a	+	b

gen	=	fibonacci()
print([next(gen)	for	_	in	range(8)])
#	[0,	1,	1,	2,	3,	5,	8,	13]
```

##Decoradores
Um decorador é uma função que recebe outra função e retorna uma versão modificada dela, permitindo adicionar comportamento sem alterar o codigo original

```python
import time
from functools import wraps

def medir_tempo(func):
  @wraps(func)
  def wrapper(*args, **kwargs):
      inicio = time.perf_counter()
      resultado = func(*args, **kwargs)
      fim = time.perf_counter()
      print(f"{func.__name__} levou {fim - inicio:.4f}s")
      return resultado
  return wrapper # This line was incorrectly indented

@medir_tempo
def tarefa_lenta():
    time.sleep(1)

tarefa_lenta()
```

##Programação Funcional

Programação funcional funciona através de funções de ordem superior nativas, favorecendo imutabilidade e funções puras, essas que sempre	retornam	o	mesmo	resultado	para	os	mesmos	argumentos,	sem	alterar
estado externo

```python
numeros = [1,2,3,4,5,6]

# o map aplica uma função a cada elemento
dobrados = list(map(lambda x: x * 2, numeros))
print(dobrados)

#filter filtra os elementos que satisfazem uma condição
pares = list(filter(lambda x: x % 2 == 0, numeros))
print(pares)

#reduce: acumula os elementos em um único valor
from functools import reduce
soma = reduce(lambda acc, x: acc + x, numeros, 0)
print(soma)
```

## Persistencia de dados
Persistir dados significa salva-los de forma durável, permitindo repercução após o encerramento do programa, as formas mais comuns em python incluem arquivos de texto, json, csv e bancos de dados

```python
# um arquivo de texto comum, salvo como txt com a mensagem escrita "olá mundo"
with open("dados.txt", "w", encoding="utf-8") as f:
  f.write("Olá, mundo!")

# mostrando o arquivo em tela
with open("dados.txt", "r", encoding="utf-8") as f:
  print(f.read())

#lendo dados
with open("dados.txt", "r") as f:
  print(f.read())
```

Salvando arquivo json

```python
#importando arquivo json
import json

dados = {"nome": "luiz", "idade": 30}

#salvar
with open("dados.json", "w", encoding="utf-8") as f:
  json.dump(dados, f)

#ler
with open("dados.json", "r", encoding="utf-8") as f:
  dados = json.load(f)
  print(dados)
```

salvando arquivo csv (planilha)

```python
import csv

with open("dados.csv", "w", encoding="utf-8") as f:
  escritor = csv.writer(f)
  escritor.writerow(["nome", "idade"])
  escritor.writerow(["luiz", 30])
```

A função `pickle` serializa qualquer objeto python (não só dados simples), mas funciona entre programas python

```python
import pickle

objeto = {"nome": "marcelo", "idade": 25}

with open("dados.pkl", "wb") as f:
  pickle.dump(objeto, f)

with open("dados.pkl", "rb") as f:
  objeto_recuperado = pickle.load(f)
  print(objeto)
```
