## Funções

uma função comum, é um bloco de codigo que executam uma ação, que podem ser usados em uma outra parte do codigo, otimizando o sistema


````python

def saudacao():
  print("olá mundo")

saudacao()
````

já as funções **lambda**, são funções anonimas que são escritas em uma linha sem ter a necessidade de usar a palavra chave def, são uteis para escrever


````python
# operações simples
saudacao = lambda: print("ola mundo")
saudacao()
#pode ser usada para operações matemáticas, já com a atribuição das variáveis de
#dentro de uma função
soma = lambda x, y: x + y
print(soma(1, 1))
````

Tratamento de Erros

````python
#TRY e EXCEPT são funções que servem para tratar erros, vamos usar o codigo anterior como
#exemplo
try:
  soma = lambda a, b: a + b
  print(soma(1, 1))
except TypeError:
  print("erro na execução do codigo")
````

Listas

As listas são usadas para armazenar multiplos dados, elas podem ser criadas de forrma **implícita**, listando os elementos entre colchetes.

````python
frutas = ['banana', 'limão', 'abacaxi', 'maça'] # Listas que declaram Strings
numeros = [21,33,1,25,67,3] # Listas que declaram numeros
dados = ['carlos', 20, True, 1.4] # Listas que declaram qualquer tipo de dados
print(frutas)
print(numeros)
print(dados)
````

````python
# Elas podem ser criadas de forma explicita usando a função "List"
# listando de 1 a 5, as posições de uma lista sempre começarão com 0
print('posições em memoria')
print('1 ','2 ','3 ','4 ','5 ')
list(range(5))
````

````python
frutas = list(('banana', 'limão', 'abacaxi', 'maça'))
print(frutas)
````

````python
# selecionando elemento dentro de uma lista por sua posição de memoria
frutas[0]
````

````python
print(frutas[1]) # selecionando elemento dentro de uma lista usando print
````

````python
# é possivel selecionar intervalos entre elementos em tela
print(frutas[2:4])
# e selecionar os primeiros e os ultimos elementos
print(frutas[:2])
print(frutas[-2:])
````

````python
# para alterar um valor de uma lista
frutas[1] = 'laranja'
print(frutas)
# para alterar diversos valores de uma lista
frutas[1:3] = ['uva', 'pera']
print(frutas)
````

````python
# para determinar o tamanho de uma lista, é usado a função len
print(len(frutas))
````

````python
# para verificar se tem um elemento em uma lista, é usado a função in
print("banana" in frutas)
````

````python
# para inserir dados em uma lista, é usado append
frutas.append("abacate")
print(frutas)
````

````python
# para inserir um elemento em uma lista em uma posição especifica, é usado insert
frutas.insert(4, 'tomate')
print(frutas)
````

Tuplas

As tuplas são usadas para armezenar multiplos itens em uma variável, ao contrário das listas, as tuplas são **imutáveis**, ou seja não alteráveis e ordenados sem uma ordem especifica que não muda, além de ser permitido valores duplicados.

````python
# tipos de dados que podem ser usados em uma tupla
nomes = ('lucas', 'fernando', 'gilmar', 'cesar')
numeros = (10, 20, 30, 40)
booleano = (True, False, True)
print('string >> ', nomes)
print('number >> ', numeros)
print('boolean >> ', booleano)
````

````python
# é possivel usar o construtor tuple para formar uma tupla
frutas = tuple(('maça', 'laranja', 'limão')) # Note os dois parenteses a mais
print(frutas)
````

````python
# para definir o tamanho de uma tupla, é usado a função len
print(len(frutas)) #frutas 1 possúi 3 valores
````

````python
# é possivel criar uma tupla com um único item
tupla_unica = ('melão',) # essa que se caracteriza com uma virgula após a sua atribuição
print('Tipo de dado: ', type(tupla_unica))
tupla_unicar = ('melão') # esse exemplo não seria uma tupla, e sim uma string
print('Tipo de dado: ', type(tupla_unicar))
````

Sets

Um set é uma coleção que armazena dados não ordenados, imutáveis e **não** permite valores duplicados

````python
# nota-se que o cojunto criado teve as posições dos dados trocados, pois um set
# armazena dados não ordenados
frutas = {'maça', 'laranja', 'limão'}
print(frutas)
````

````python
# nesse exemplo, o dado que foi repetido foi mostrado apenas 1 vez
frutas_rep = {'maça', 'maça', 'laranja', 'limão'}
print(frutas_rep)
````

Dicionários

Os dicionários são usados para armazenar valores de dados em pares, tipo CHAVE: VALOR, um dicionário é uma coleção que é ordenada, alteravel e que não permite duplicados, semelhante a uma tabela de um banco de dados.

````python
# criando dois dicionarios com atributos diferentes
banana = {
    'nome': 'banana',
    'cor': 'amarelo',
    'origem': 'asia',
    'tem_semente': False
}

coco = {
    'nome': 'coco',
    'cor': 'verde',
    'origem': 'asia',
    'tem_semente': False
}

print(banana)
print(coco)
````
